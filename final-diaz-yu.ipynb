```python

def make_audio_chunks(track_list, chunk_size):
    """
    splits audio into evenly sized chunks. Used to increase the number of examples in set
    
    parameters
    
    track_list: a list of GuitarSet objects
    chunk_size: desired length of chunk, in samples
    
    
    """
    list_of_filepaths = []
    list_of_track_ids = []
    for track_id, track in track_list:
        # get audio
        track_audio, sr = librosa.load(track.audio_mic_path, sr=None)

        chunked_audio = librosa.util.frame(track_audio, chunk_size,chunk_size)

        for n in range(chunked_audio.shape[1]):
            new_track_id = track_id + "_chunk_" + str(n)
            file_path = "Audio/" + new_track_id + ".wav"
            list_of_filepaths.append(file_path)
            list_of_track_ids.append(new_track_id)
            
    
    return list_of_filepaths, list_of_track_ids
```

```python
def getTimeConstant(scalar, time_ms, sr):
    """
    gets the coefficient used for compressor peak detector
    
    parameters
    
    scalar: sets the target timing behavior. e.g. -1 means coeff will refer to time it takes
            for the smoothed signal to reach roughly 63.3% of its final value.
    time_ms: the desired time constant, in milliseconds
    sr: sample rate
    
    """
    return np.exp(scalar / (time_ms * .001 * sr))

    aTau = getTimeConstant(-1.0, attack, sr)
    rTau = getTimeConstant(-1.0, release, sr)

def compress(input_audio, threshold=-40.0, makeup=0.0, knee=0.0, ratio=8.0, aTau=-1, rTau=-1):
    
    """

    Takes a mono audio signal, compresses it, and returns a vector of compressed audio.
    
    
    parameters
    
    input_audio: take an (1,n) np.array of audio samples
    threshold: the signal level above which compression will begin, in decibels
    makeup: gain applied after compression, in decibels
    knee: creates a transition area around threshold where compressor will start
          acting, at an increasing ratio, until it has cleared the knee, at which point
          it will act at the full ratio
    ratio: indicates how much compressor will act on signal past the threshold
    aTau/rTau: coefficients for the peak detector. They can be found using getTimeConstant
    
    """
    
    blockSize = len(input_audio)
    control_signal = np.zeros(blockSize)
    
    y_l_prev = 0.0
    
    for n in range (blockSize):
        
        #x_g is the input to the gain computer
        x_g = abs(input_audio[n])
    
        #convert to decibels
        if x_g < 0.0000001:
            x_g = -120.0
        else:
            x_g = 20.0 * np.log10(x_g)
            
        #compute gain
        overshoot = x_g - threshold
        
        if knee > 0.0:
            
            doubleOvershoot = 2.0 * overshoot
            
            if  doubleOvershoot < -knee:
                y_g = x_g
            elif abs(doubleOvershoot) <= knee:
                y_g = x_g + ((1.0 / ratio - 1.0) * ((overshoot + knee / 2.0) ** 2) / (2.0 * knee))
            else:
                y_g = threshold + overshoot / ratio
        else:
            
            if x_g > threshold:
                y_g = threshold + overshoot / ratio
            else:
                y_g = x_g
    

        #x_l is the input to the level detector
        x_l = x_g - y_g
        

        #choose timing coefficient
        if x_l > y_l_prev:
            coeff = aTau
        else: 
            coeff = rTau
    
        #smooth control signal sample
        y_l = coeff * (y_l_prev - x_l) + x_l
    
        #store output
        y_l_prev = y_l
        
        #convert to linear
        control_signal[n] = pow(10.0, ((makeup - y_l)  / 20.0)) 
    
    #apply gain to signal
    compressed_audio = input_audio * control_signal
    return compressed_audio
    
```

```python
def get_stats(features):
    """Compute summary statistics (mean and standard deviation) over a matrix of features.
    Make sure the statistics are computed across features
    (i.e. over all examples, compute the mean of each feature)

    Parameters
    ----------
    features: np.array (n_examples, n_features)
        Matrix of features

    Returns
    -------
    features_mean: np.array (n_features)
        The mean of the features
    features_std: np.array (n_features)
        The standard deviation of the features

    """
    features_mean = np.mean(features, axis=0)
    features_std = np.std(features, axis=0)
    
    return features_mean, features_std
    
    
def normalize(features, features_mean, features_std):
    """Normalize (standardize) a set of features using the given mean and standard deviation

    Parameters
    ----------
    features: np.array (n_examples, n_features)
        Matrix of features
    features_mean: np.array (n_features)
        The mean of the features
    features_std: np.array (n_features)
        The standard deviation of the features

    Returns
    -------
    features_norm: np.array (n_examples, n_features)
        Standardized features

    """
    
    return (features - features_mean) / features_std
```

```python
def get_crest_feature(input_audio, frame_length=2048, hop_length=2048):
    """
    Get the crest factor for an audio signal, split into frames
    
    
    Returns
    -------
    features: np.array (1, 2)
        The features, which should be the mean and standard deviation of the crest factor over time
        stacked into a single vector.
    
    """
    
    framed_input = librosa.util.frame(input_audio, frame_length,hop_length)
        
    vec_len = framed_input.shape[1]

    crest_values = np.zeros(vec_len)

    for n in range(vec_len):
        peak_value = np.max(abs(framed_input[:,n]))
        rms_value = ((np.sum(framed_input[:,n] ** 2)) / frame_length) ** .5
    
        crest_values[n] = 20.0 * np.log10(peak_value / rms_value)

    crest_mean, crest_std = get_stats(crest_values)

    return np.array([crest_mean, crest_std])

```

```python
import librosa

def compute_mfccs(y, sr, n_fft=2048, hop_length=512, n_mels=128, n_mfcc=20):
    """Compute mfccs for an audio file, removing the 0th MFCC coefficient
    
    Parameters
    ----------
    y : np.array
        Mono audio signal
    sr : int
        Audio sample rate
    n_fft : int
        Number of points for computing the fft
    hop_length : int
        Number of samples to advance between frames
    n_mels : int
        Number of mel frequency bands to use
    n_mfcc : int
        Number of mfcc's to compute
    
    Returns
    -------
    mfccs: np.array (t, n_mfcc - 1)
        Matrix of mfccs

    """
    mfcc_matrix = librosa.feature.mfcc(y, sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels, n_mfcc=n_mfcc)
    
    return mfcc_matrix[1:,:].T


def compute_features(audio, sr, n_fft=2048, hop_length=512, n_mels=128, n_mfcc=20):
    """Compute summary statistics for MFCCs and Crest Factor over time, given an audio file.

    Parameters
    ----------
    audio : np.array
        Mono audio signal
    sr : int
        Audio sample rate
    n_fft : int
        Number of points for computing the fft
    hop_length : int
        Number of samples to advance between frames
    n_mels : int
        Number of mel frequency bands to use
    n_mfcc : int
        Number of mfcc's to compute

    Returns
    -------
    features: np.array (1, 2*(n_mfcc - 1) + 2)
        The features, which should be the mean and standard deviation of the MFCCs and the 
        crest stacked into a single vector

    """
    mfcc_matrix = compute_mfccs(audio, sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels, n_mfcc=n_mfcc)
    mfcc_mean, mfcc_std = get_stats(mfcc_matrix)
    
    #get the crest feature here so the matrix does it's thing
    crest_feature = get_crest_feature(audio)
    
    return np.concatenate((mfcc_mean, mfcc_std, crest_feature), axis=0).reshape(1,-1)
    #return np.concatenate((mfcc_mean, mfcc_std), axis=0).reshape(1,-1)

def get_features_and_labels(track_dict, n_fft=2048, hop_length=512, n_mels=128, n_mfcc=20):
    
    """get mfcc and crest factor features as well as labels

    Parameters
    ----------
    track_dict : dict
        dict of augmented audio file paths
    n_fft : int
        Number of points for computing the fft
    hop_length : int
        Number of samples to advance between frames
    n_mels : int
        Number of mel frequency bands to use
    n_mfcc : int
        Number of mfcc's to compute

    Returns
    -------
    feature_matrix: np.array (len(track_list), 2*(n_mfcc - 1)+2)
        The features for each track, stacked into a matrix.
    """
    num_file = len(track_dict.keys())
    label_array = []
    feature_matrix = np.zeros([num_file, 2*(n_mfcc - 1)+2] )
    
    n = 0
    for track in track_dict.keys():
        audio, sr = librosa.load(track_dict[track]['filepath'], sr=None)
        feature_matrix[n,:] = compute_features(audio, sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels, n_mfcc=n_mfcc)
        label_array.append(track_dict[track]['ratio'])    
        n+=1
        
    return feature_matrix, label_array
```

```python
def fit_knn_clf(train_features, train_labels, validation_features, validation_labels):
    """Fit a k-nearest neighbor classifier and choose the k which maximizes the
    *f-measure* on the validation set.
    
    Plot the ba-score on the validation set as a function of k.

    Parameters
    ----------
    train_features : np.array (n_train_examples, n_features)
        training feature matrix
    train_labels : np.array (n_train_examples)
        training label array
    validation_features : np.array (n_validation_examples, n_features)
        validation feature matrix
    validation_labels : np.array (n_validation_examples)
        validation label array

    Returns
    -------
    knn_clf : scikit learn classifier
        Trained k-nearest neighbor classifier with the best k
    best_k : int
        The k which gave the best performance
    """
    
    k_values = [1,5,10,35,45,50,75,100]

    knn_clf = []
    predicted_labels = []
    ba_score = []

    end = -1

    for n in k_values:
        knn_clf.append(KNeighborsClassifier(n_neighbors=n))
        knn_clf[end].fit(train_features, train_labels)
        predicted_labels.append(knn_clf[end].predict(validation_features))
        ba_score.append(sklearn.metrics.balanced_accuracy_score(validation_labels, predicted_labels[end]))
        print("ba score for k = %i: "%n + "%f"%ba_score[end])
    
    best_k_index = np.argmax(ba_score)
    
    plt.figure(figsize=(15, 7))

    plt.title("balanced accuracy score on Validation Data for Different k values")
    plt.xlabel("k")
    plt.ylabel("balanced accuracy score")
    plt.scatter(np.arange(len(k_values)), ba_score)
    plt.xticks(np.arange(len(k_values)), k_values)
    plt.show()
    
    return knn_clf[best_k_index], k_values[best_k_index]
   
```

```python
def plot_confusion_matrix(true_labels, predicted_labels, title, classes):

    cm = confusion_matrix(true_labels, predicted_labels)
    cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    
    plt.figure()
    fig, ax = plt.subplots(figsize=(15,7))

    im = ax.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    ax.figure.colorbar(im, ax=ax)

    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           xticklabels=classes, yticklabels=classes,
           title=title,
           ylabel='True label',
           xlabel='Predicted label')
    ax.axis('tight')

    plt.show()
    
    return cm
```
